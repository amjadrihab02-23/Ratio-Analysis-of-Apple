```python

import pandas as pd
import yfinance as yf
symbol = "AAPL"
company = yf.Ticker(symbol)

bs = company.balance_sheet

try:
    current_assets = bs.loc["Current Assets"].iloc[0]
    current_liabilities = bs.loc["Current Liabilities"].iloc[0]
    total_debt = bs.loc["Total Debt"].iloc[0]
    total_equity = bs.loc["Stockholders Equity"].iloc[0]

    current_ratio = current_assets / current_liabilities
    debt_to_equity_ratio = total_debt / total_equity


    print(f"-- Analysis for {symbol} ----")
    print(f"Current Ratio: {current_ratio:.2f}")
    print(f"Debt to Equity Ratio: {debt_to_equity_ratio:.2f}")
    if current_ratio <1.5:
        print("Warning: Current Ratio is below the recommended threshold of 1.5.")
    else:
        print("Current Ratio is within the recommended range.")
    if debt_to_equity_ratio >2:
        print("Warning: Debt to Equity Ratio is above the recommended threshold of 2.")
    else:
        print("Debt to Equity Ratio is within the recommended range.")

except KeyError as e:
    print(f"Data not available for {e} in the balance sheet.")
```
